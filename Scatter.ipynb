```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
CYCLES = 10
ocupacao_media = sorted(np.random.exponential(size=10))[::-1]
ocupacao_media_lib = sorted(np.random.exponential(size=10))[::-1]
ocupacao_media_genetico = sorted(np.random.exponential(size=10))[::-1]
```

```python
X = list(range(1, CYCLES + 1))

fig, ax = plt.subplots()

ax.plot(X, ocupacao_media, 'r:')
ax.plot(X, ocupacao_media_lib, 'b-.')
ax.plot(X, ocupacao_media_genetico, 'g-')

ax.legend(['Cultural', 'Networkx', 'Genético'])

plt.show()
```
