```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import pprint
```

```python
G = nx.read_gml("Geant2012.gml")
```

```python
G.nodes
```

```python
pos = nx.spring_layout(G)
nx.draw_networkx_labels(G, pos)
nx.draw(G, pos)
plt.show()
```

```python
dijkstra_distance = {node: {} for node in G.nodes}
for i in G.nodes:
    for j in G.nodes:
        if i == j:
            continue
        dijkstra_distance[i][j] = list(nx.all_shortest_paths(G, i, j))
```

```python
dijkstra_distance
```

```python
for source in dijkstra_distance:
    print("Source:", source)
    for target in dijkstra_distance[source]:
        print("Target:", target)
        pprint.pprint([path for path in dijkstra_distance[source][target]])
```
